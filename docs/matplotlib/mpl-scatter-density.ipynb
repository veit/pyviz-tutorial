# mpl-scatter-density

[mpl-scatter-density](https://github.com/astrofrog/mpl-scatter-density) facilitates the creation of interactive scatter density maps from large amounts of data.

## Installation

``` bash
$ pipenv install mpl-scatter-density
```
This automatically installs the dependencies [Numpy](https://numpy.org/), [Matplotlib](../matplotlib/index.rst) and [fast-histogram](https://github.com/astrofrog/fast-histogram).

## Intended use

1. Imports

````python
import matplotlib.pyplot as plt
import mpl_scatter_density
import numpy as np
````

2. Generate sample data

````python
N = 10000000
x = np.random.normal(4, 2, N)
y = np.random.normal(3, 1, N)
````

3. There are essentially two uses for mpl-scatter-density:

   1. `projection='scatter_density'`
   2. `ScatterDensityArtist`

3. 1. `projection='scatter_density'`

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection="scatter_density")
ax.scatter_density(x, y)
ax.set_xlim(-5, 10)
ax.set_ylim(-5, 10)
fig.savefig("gaussian.png")
````

3. 2. `ScatterDensityArtist`

`ScatterDensityArtist` is also used in the previous example, but without specifying it directly. In the following example, we add it explicitly to the axes:

````python
from mpl_scatter_density import ScatterDensityArtist


a = ScatterDensityArtist(ax, x, y)
ax.add_artist(a)
````
