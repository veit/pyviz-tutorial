# seaborn example

1. Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

2. Define style

   [seaborn.set](https://seaborn.pydata.org/generated/seaborn.set.html) is an alias for [seaborn.set_theme](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme), which can be used to define different styles, palettes, fonts and colours, for example:

```python
sns.set(style="white", color_codes=True)
```

3. Creating a bivariate random data set

   For this we use [numpy.random.RandomState.multivariate_normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.RandomState.multivariate_normal.html) from [numpy.random.RandomState](https://numpy.org/doc/stable/reference/random/legacy.html), where a seed of `9`, `0` as mean values and 100 samples result in a not quite symmetrical covariance matrix:

```python
rs = np.random.RandomState(9)
mean = [0, 0]
cov = [(1, 0), (0, 2)]
x, y = rs.multivariate_normal(mean, cov, 100).T
```

4. `JointGrid` for bivariate plots

   Bivariate plots can be generated with [seaborn.JointGrid](https://seaborn.pydata.org/generated/seaborn.JointGrid.html). If both functions pass different keyword arguments, [JointGrid.plot_joint()](https://seaborn.pydata.org/generated/seaborn.JointGrid.plot_joint.html#seaborn.JointGrid.plot_joint) and [JointGrid.plot_marginals()](https://seaborn.pydata.org/generated/seaborn.JointGrid.plot_marginals.html#seaborn.JointGrid.plot_marginals) must be used:

```python
grid = sns.JointGrid(x=x, y=y, space=0, height=6, ratio=10)
grid.plot_joint(plt.scatter, color="g")
grid.plot_marginals(sns.rugplot, height=0.2, color="g")
```
