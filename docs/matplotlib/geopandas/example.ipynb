# GeoPandas examples

In the following examples we use the [nybb](https://data.cityofnewyork.us/dataset/nybb/7t3b-ywvw) dataset.

## GeoPandas-Plot

1. Import

```python
import geopandas as gpd

from geodatasets import get_path
```

2. Read data

```python
nybb = get_path("nybb")
```

```python
df = gpd.read_file(nybb)
```

3. Display the first lines

```python
df.head()
```

4. Plotting

```python
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor="k")
```

`figsize` specifies the size of the plot, `alpha` the transparency and `edgecolor` the edge colour.

## Plotting in Folium

[Folium](https://python-visualization.github.io/folium/), or more precisely the [leaflet.js](https://leafletjs.com/) uses the values for latitude and longitude by default. We therefore have to convert our values first:

1. Determine coordinates

   To be able to use the data with [EPSG codes](https://en.wikipedia.org/wiki/EPSG_Geodetic_Parameter_Dataset), GeoPandas offers [geopandas.GeoDataFrame.crs](https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.crs.html) ([spatial reference system](https://en.wikipedia.org/wiki/Spatial_reference_system)).

```python
print(df.crs)
```

```python
df = df.to_crs(epsg=4326)
print(df.crs)
df.head()
```

```python
import folium
```

```python
m = folium.Map(
    location=[40.70, -73.94], zoom_start=10, tiles="CartoDB positron"
)
for _, r in df.iterrows():
    # without simplifying the representation of each borough, the map might not be displayed
    sim_geo = gpd.GeoSeries(r["geometry"]).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(
        data=geo_j,
    )
    folium.Popup(r["BoroName"]).add_to(geo_j)
    geo_j.add_to(m)
m
```
